```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Define a function
## y = f(x) = 5x + 0.5

```python
def f(x):
    return 5*x + 0.5
```

## Method 1: Iteration

```python
x_list = range(10)
for x in x_list:
    print(f(x))
```

## Method 2: List appending

```python
x_list = range(10)
y_list = []
for x in x_list:
    y = f(x)
    y_list.append(y)
print(y_list)
```

## Method 3: List comprehension

```python
x_list = range(10)
y_list = [f(x) for x in x_list]
print(y_list)
```

## Method 4: Map

```python
x_list = range(10)
y_list = map(f, x_list)
print(y_list)
```

## Method 5: Numpy Arrays

```python
x_list = np.arange(10)
y_list = f(x_list)
print(y_list)
```

# PLOT: X and Y

```python
def f(x):
    return ((x**2) / 10.0) + 2.5

x = np.arange(1000)
y = f(x)

plt.figure()
plt.plot(x, y)
plt.show()
```
